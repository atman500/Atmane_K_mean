```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Combine X and y to drop rows with NaN values consistently
df_combined = pd.concat([X, y], axis=1)
df_combined.dropna(inplace=True)

X_cleaned = df_combined[X.columns]
y_cleaned = df_combined[y.name]

# Initialize the Linear Regression model
model = LinearRegression()

# Fit the model using the entire dataset (as implied by 'classic model' without training split)
# If a train-test split is desired for evaluating performance, that would be the next step.
model.fit(X_cleaned, y_cleaned)

# Display the coefficients and intercept
print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)

# Make predictions on the training data
y_pred = model.predict(X_cleaned)

# Evaluate the model
mse = mean_squared_error(y_cleaned, y_pred)
r2 = r2_score(y_cleaned, y_pred)

print(f"\nMean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2): {r2:.2f}")
```
